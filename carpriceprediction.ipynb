```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
car_data = pd.read_csv('/content/car data.csv')
```

```python
car_data.head()
```

```python
car_data.shape
```

```python
car_data.describe()
```

```python
car_data.info()
```

```python
car_data.isnull().sum()
```

```python
print(car_data.Fuel_Type.value_counts())
print(car_data.Seller_Type.value_counts())
print(car_data.Transmission.value_counts())
```

```python
car_data.replace({'Fuel_Type': {'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
car_data.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
car_data.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_data.head()
```

```python
X =car_data.drop(['Selling_Price','Car_Name'],axis=1)
Y = car_data['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train,Y_train)
```

```python
train_data = lin_reg.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, train_data)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train,train_data)
plt.xlabel('Actual data')
plt.ylabel('Predicted price')
plt.title('Actual price VS Predicted price')
plt.show()
```

```python
test_data = lin_reg.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test,test_data)
plt.xlabel('Actual data')
plt.ylabel('Predicted price')
plt.title('Actual price VS Predicted price')
plt.show()
```

```python
#  implementing through lasso Regression
```

```python
lasso_reg = Lasso()
```

```python
lasso_reg.fit(X_train,Y_train)
```

```python
training_data_prediction = lasso_reg.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
test_data_prediction = lasso_reg.predict(X_test)
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
