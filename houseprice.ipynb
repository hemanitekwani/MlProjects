```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df = pd.read_csv('/content/Boston.csv')
```

```python
X =df.drop(['medv'], axis=1)
Y = df['medv']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
tranning_data = model.predict(X_train)
```

```python
print(tranning_data)
```

```python
tranning_data.shape
```

```python
accuracy_prediction=metrics.r2_score(tranning_data,Y_train)
print(accuracy_prediction)
```

```python
score_2 = metrics.mean_absolute_error(Y_train,tranning_data)
print(score_2 )
```

```python
plt.scatter(tranning_data,Y_train)
plt.xlabel("Actual prices")
plt.xlabel("Predicted prices")
plt.title("Actual prices Vs Predicted prices" )
plt.show()
```

```python
test_data =model.predict(X_test)
score_1 = metrics.r2_score(Y_test, test_data)
score_2 =metrics.mean_absolute_error(Y_test, test_data)
print(score_1)
print(score_2)
```
